```python
 # Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary
- - -
* Calculate the total number of schools
* Calculate the total number of students
* Calculate the total budget
* Calculate the average math score 
* Calculate the average reading score
* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
* Calculate the percentage of students with a passing math score (70 or greater)
* Calculate the percentage of students with a passing reading score (70 or greater)
* Create a dataframe to hold the above results
* Optional: give the displayed data cleaner formatting

```python
number_schools=school_data_complete["school_name"].nunique()
number_students=school_data_complete["student_name"].count()
total_budget="${:0,.0f}".format(school_data_complete["budget"].unique().sum())
average_math=round(school_data_complete["math_score"].mean(),2)
average_reading=round(school_data_complete["reading_score"].mean(),2)
passing_rate=(average_math+average_reading)/2
pass_math=round((school_data_complete[school_data_complete["math_score"]>=70].count()["math_score"])/number_students,2)
pass_reading=round((school_data_complete[school_data_complete["reading_score"]>=70].count()["reading_score"])/number_students,2)
```

```python
e={
    "Number of schools":number_schools,"Number of students":"{:,}".format(number_students),"Total budget":total_budget,
    "Average math score":average_math,"Average reading score":average_reading,"Overall passing rate":passing_rate,
    "% passing math":pass_math,"% passing reading":pass_reading
}
district_s=pd.DataFrame(data=e,index=[0])
district_s
```

## School Summary
- - - 
Create an overview table that summarizes key metrics about each school, including:
* School Name
* School Type
* Total Students
* Total School Budget
* Per Student Budget
* Average Math Score
* Average Reading Score
* % Passing Math
* % Passing Reading
* Overall Passing Rate (Average of the above two)

```python
school_data_grouped=school_data_complete.groupby(["school_name","type"])
students_by_school=school_data_grouped["student_name"].count()
budget_by_school=round(school_data_grouped["budget"].mean(),1)
budget_per_student=round(budget_by_school/students_by_school,2)
average_math_by_school=round(school_data_grouped["math_score"].mean(),2)
average_reading_by_school=round(school_data_grouped["reading_score"].mean(),2)
filtered_df1=school_data_complete.loc[school_data_complete["math_score"]>=70,:]
pass_math_by_school=round(filtered_df1.groupby(["school_name","type"])["math_score"].count()*100/students_by_school,2)
filtered_df2=school_data_complete.loc[school_data_complete["reading_score"]>=70,:]
pass_reading_by_school=round(filtered_df2.groupby(["school_name","type"])["reading_score"].count()*100/students_by_school,2)
passing_rate_by_school=(pass_math_by_school+pass_reading_by_school)/2
```

## Top Performing Schools (By Passing Rate)

```python
# Build a dataframe that diplays the statistics by school, and sort it to generate the top performing schools
f={
    "Number of students":students_by_school,"Total school budget":budget_by_school,"Budget per student":budget_per_student,
    "Average math score":average_math_by_school,"Average reading score":average_reading_by_school,
    "% passing math":pass_math_by_school,"% passing reading":pass_reading_by_school,"Overall passing rate":passing_rate_by_school
}
school_summary=pd.DataFrame(data=f)
school_summary=school_summary.sort_values(["Overall passing rate"],ascending=False)
school_summary.head(5)
```

```python
# Create a series for each grade with the average score per school
filtered_9th=school_data_complete.loc[school_data_complete["grade"]=='9th',:]
math_score_9th=round(filtered_9th.groupby(["school_name"])["math_score"].mean(),2)
filtered_10th=school_data_complete.loc[school_data_complete["grade"]=='10th',:]
math_score_10th=round(filtered_10th.groupby(["school_name"])["math_score"].mean(),2)
filtered_11th=school_data_complete.loc[school_data_complete["grade"]=='11th',:]
math_score_11th=round(filtered_11th.groupby(["school_name"])["math_score"].mean(),2)
filtered_12th=school_data_complete.loc[school_data_complete["grade"]=='12th',:]
math_score_12th=round(filtered_12th.groupby(["school_name"])["math_score"].mean(),2)
```

```python
# Build a dataframe that diplays the average score by school and by grade
f={
    "9th":math_score_9th,"10th":math_score_10th,"11th":math_score_11th,"12th":math_score_12th
}
math_score_by_grade=pd.DataFrame(data=f)
math_score_by_grade
```

## Scores by School Spending
- - - 
Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
* Average Math Score
* Average Reading Score
* % Passing Math
* % Passing Reading
* Overall Passing Rate (Average of the above two)

```python
school_summary.describe()
```

```python
# define bins
budget_bins = [0, 585, 625, 650, 1000]
group_names = ["<$585", "$585-625", "$625-650", ">$650"]
# Cut budget by bins
school_summary["Spending range"]=pd.cut(school_summary["Budget per student"],budget_bins,labels=group_names)
school_summary
```

```python
# group schools by budget range and estimate statistics (we will compare at school level)
school_budget_grouped=school_summary.groupby("Spending range")
students_by_budget=school_budget_grouped["Number of students"].sum()
average_math_by_budget=round(school_budget_grouped["Average math score"].mean(),2)
average_reading_by_budget=round(school_budget_grouped["Average reading score"].mean(),2)
pass_math_by_budget=round(school_budget_grouped["% passing math"].mean(),2)
pass_reading_by_budget=round(school_budget_grouped["% passing reading"].mean(),2)
passing_rate_by_budget=(pass_math_by_budget+pass_reading_by_budget)/2
```

```python
# Build a dataframe that diplays the statistics by range of budget
f={
    "Average math score":average_math_by_budget,"Average reading score":average_reading_by_budget,
    "% passing math":pass_math_by_budget,"% passing reading":pass_reading_by_budget,"Overall passing rate":passing_rate_by_budget
}
Budget_range_summary=pd.DataFrame(data=f)
Budget_range_summary
```
